## Q1 X값 생성

- 전체 112 신고건과 절도, 폭력 관련 신고건을 count하여 x값으로 사용

### 1. 전체

```python
## 신고 접수건수 total count
report_total_cnt = report_112.groupby(['jur_stn','date']).count()[['report_sx']]
report_total_cnt.reset_index(inplace=True)
report_total_cnt = report_total_cnt.rename(columns={'report_sx' : 'report_total_cnt'})
report_total_cnt.head()
```

```python
report_total_cnt['report_total_cnt'] = report_total_cnt.apply(lambda x : x['report_total_cnt'] * 5 if x['date'] =='2018(상반기)' else x['report_total_cnt'] , axis= 1   )
report_total_cnt.head(3)
```

```python
# 전체 신고자 성별 count
total_sx_cnt = report_112.groupby(['jur_stn','date', 'report_sx']).count()[['year']]
total_sx_cnt.reset_index(inplace=True)
total_sx_cnt = total_sx_cnt.rename(columns={'year' : 'total_sx_cnt'})
total_sx_cnt.head()
```

```python
total_sx_cnt['total_sx_cnt'] = total_sx_cnt.apply(lambda x : x['total_sx_cnt'] * 5 if x['date'] =='2018(상반기)' else x['total_sx_cnt'] , axis= 1   )
total_sx_cnt.head(3)
```

```python
## X값으로 만들기 위해 신고자 성별 값을 column으로 만들고 count 값을 각 row에 삽입한다
temp_data = total_sx_cnt.drop_duplicates(subset=['jur_stn','date'])[['jur_stn','date']]
temp_data= temp_data.reset_index()
del temp_data['index']

for i in total_sx_cnt['report_sx'].unique() :
    temp_data['report_sx_'+i] = 0
    
for i in range(len(total_sx_cnt)) :
    temp_name = total_sx_cnt['report_sx'].iloc[i]
    temp = temp_data[(temp_data['jur_stn']==total_sx_cnt['jur_stn'].iloc[i])&(temp_data['date']==total_sx_cnt['date'].iloc[i])]
    temp[temp_name] = total_sx_cnt['total_sx_cnt'].iloc[i]
    temp_data['report_sx_'+temp_name].iloc[temp.index] = total_sx_cnt['total_sx_cnt'].iloc[i]
total_sx_cnt_df = temp_data
total_sx_cnt_df.head(3)
```

```python
# total cnt 와 성별 cnt column merge
temp_total = pd.merge(report_total_cnt, total_sx_cnt_df)
temp_total.head()
```

### 2. 폭력

```python
# 신고 건수 중 폭력 건만 total count 
vio_total_cnt = report_112[report_112['inc_info'].str.contains('폭력')].groupby(['jur_stn','date']).count()[['report_sx']]
vio_total_cnt.reset_index(inplace=True)
vio_total_cnt = vio_total_cnt.rename(columns={'report_sx' : 'vio_total_cnt'})
vio_total_cnt.head()
```

```python
vio_total_cnt['vio_total_cnt'] = vio_total_cnt.apply(lambda x : x['vio_total_cnt'] * 5 if x['date'] =='2018(상반기)' else x['vio_total_cnt'] , axis= 1   )
vio_total_cnt.head(3)
```

```python
# 폭력 건 신고자 성별 count
vio_total_sx_cnt =report_112[report_112['inc_info'].str.contains('폭력')].groupby(['jur_stn','date', 'report_sx']).count()[['year']]
vio_total_sx_cnt.reset_index(inplace=True)
vio_total_sx_cnt = vio_total_sx_cnt.rename(columns={'year' : 'vio_total_sx_cnt'})
vio_total_sx_cnt.head()
```

```python
vio_total_sx_cnt['vio_total_cnt'] = vio_total_sx_cnt.apply(lambda x : x['vio_total_sx_cnt'] * 5 if x['date'] =='2018(상반기)' else x['vio_total_sx_cnt'] , axis= 1   )
vio_total_sx_cnt.head(3)
```

```python
## X값으로 만들기 위해 신고자 성별 값을 column으로 만들고 count 값을 각 row에 삽입한다
temp_data = vio_total_sx_cnt.drop_duplicates(subset=['jur_stn','date'])[['jur_stn','date']]
temp_data= temp_data.reset_index()
del temp_data['index']

for i in vio_total_sx_cnt['report_sx'].unique() :
    temp_data['report_sx_'+i] = 0
    
for i in range(len(vio_total_sx_cnt)) :
    temp_name = vio_total_sx_cnt['report_sx'].iloc[i]
    temp = temp_data[(temp_data['jur_stn']==vio_total_sx_cnt['jur_stn'].iloc[i])&(temp_data['date']==vio_total_sx_cnt['date'].iloc[i])]
    temp[temp_name] = vio_total_sx_cnt['vio_total_sx_cnt'].iloc[i]
    temp_data['report_sx_'+temp_name].iloc[temp.index] = vio_total_sx_cnt['vio_total_sx_cnt'].iloc[i]
vio_total_sx_cnt_df = temp_data
vio_total_sx_cnt_df.head(3)
```

```python
# 폭력 건 total cnt 와 성별 cnt column merge
temp_vio_total = pd.merge(vio_total_cnt, vio_total_sx_cnt_df)
temp_vio_total.head()
```

```python
# 폭력 카테고리 별 total count
vio_part_cnt = report_112[report_112['inc_info'].str.contains('폭력')].groupby(['jur_stn','date','inc_info']).count()[['report_sx']]
vio_part_cnt.reset_index(inplace=True)
vio_part_cnt = vio_part_cnt.rename(columns={'report_sx' : 'vio_part_cnt'})
vio_part_cnt.head()
```

```python
## X값으로 만들기 위해 폭력 카테고리 값을 column으로 만들고 count 값을 각 row에 삽입한다
temp_data = vio_part_cnt.drop_duplicates(subset=['jur_stn','date'])[['jur_stn','date']]
temp_data= temp_data.reset_index()
del temp_data['index']

for i in vio_part_cnt['inc_info'].unique() :
    temp_data['inc_info_'+i] = 0
    
for i in range(len(vio_part_cnt)) :
    temp_name = vio_part_cnt['inc_info'].iloc[i]
    temp = temp_data[(temp_data['jur_stn']==vio_part_cnt['jur_stn'].iloc[i])&(temp_data['date']==vio_part_cnt['date'].iloc[i])]
    temp[temp_name] = vio_part_cnt['vio_part_cnt'].iloc[i]
    temp_data['inc_info_'+temp_name].iloc[temp.index] = vio_part_cnt['vio_part_cnt'].iloc[i]
vio_part_cnt_df = temp_data
vio_part_cnt_df.head(3)
```

```python
# 폭력 건 total cnt 와 성별 cnt column merge
vio_total_df = pd.merge(temp_vio_total, vio_part_cnt_df)
vio_total_df.head()
```
