## X, y data

```python
# q5 data에서 X,y data 분리
X_data = q5_df.drop('overall_q5',axis=1)
y_data = q5_df['overall_q5'].dropna()
```

```python
# 2020 년도는 y값이 없으므로 제외하고 eval data로 포함
X_data = X_data.drop(X_data[X_data['date'].str.contains('2020')].index)
X_data.head()
```

```python
# 2020년 데이터로 eval data 생성
del_idx = q5_df[q5_df['overall_q5'].notnull()].index
eval_X_data = q5_df.drop(del_idx)
eval_X_data=  eval_X_data.drop('overall_q5',axis=1)
```
