```python
# act_type column 생성 (acc_case의 '-' 을 기준으로 앞은 대분류, 뒤는 소분류로 카테고리 지정)
car_acc['acc_type'] = car_acc['acc_case'].apply(lambda x : x.split('-')[0].strip())
car_acc.head(3)
```

```python
car_acc['acc_case'] = car_acc['acc_case'].apply(lambda x : x.split('-')[1].strip())
car_acc['acc_case'].value_counts()
```

## 1. 교통사고 건수 합산
- 이전 X값 만드는 방법과 유사하게 진행

### 전체

```python
# 교통사고 total cnt
total_cnt = car_acc.groupby(['jur_stn','date']).count()[['acc_case']]
total_cnt.reset_index(inplace=True)
total_cnt = total_cnt.rename(columns={'acc_case' : 'total_cnt'})
total_cnt.head(5)
```

### 가해자

```python
## 가해자 성별 cnt
ofn_sx_cnt = car_acc.groupby(['jur_stn','date','ofn_sx']).count().unstack(fill_value=0).stack()[['acc_case']]
ofn_sx_cnt.reset_index(inplace=True)
ofn_sx_cnt = ofn_sx_cnt.rename(columns={'acc_case' : 'ofn_sx_cnt'})
ofn_sx_cnt.head()
```

```python
# col 생성
df1 = ofn_sx_cnt.drop_duplicates(subset=['jur_stn','date'])[['jur_stn','date']]
df1= df1.reset_index()
del df1['index']
for i in ofn_sx_cnt['ofn_sx'].unique() :
    df1['ofn_sx_'+i] = 0
df1.head(3)
```

```python
for i in range(len(ofn_sx_cnt)) :
    sx_name = ofn_sx_cnt['ofn_sx'].iloc[i]
    temp = df1[(df1['jur_stn']==ofn_sx_cnt['jur_stn'].iloc[i])&(df1['date']==ofn_sx_cnt['date'].iloc[i])]
    df1['ofn_sx_'+sx_name].iloc[temp.index] = ofn_sx_cnt['ofn_sx_cnt'].iloc[i]
df1.head(3)
```

```python
# 성비 col 생성
df1['ofn_sx_성비'] = round(df1['ofn_sx_남'] / df1['ofn_sx_여'] * 100, 1)
df1.head(3)
```

```python
## 가해자 나이 ct
ofn_age_cnt = car_acc.groupby(['jur_stn','date','ofn_age']).count().unstack(fill_value=0).stack()[['acc_case']]
ofn_age_cnt.reset_index(inplace=True)
ofn_age_cnt = ofn_age_cnt.rename(columns={'acc_case' : 'ofn_age_cnt'})
ofn_age_cnt.head()
```

```python
ofn_age_cnt['ofn_age'].unique()
```

```python
# age 카테고리를 재분류함
age_map = { '10대' : '10대 이하', '10대 미만' : '10대 이하', '20대' : '20~30대', '30대' : '20~30대', \
        '40대' : '40~50대', '50대' : '40~50대' , '60대' : '60~70대', '70대' : '60~70대' , '80대' : '80대 이상' ,\
                                    '90대 이상' : '80대 이상', '미분류' : '미분류' }
```

```python
ofn_age_cnt['ofn_age'] = ofn_age_cnt['ofn_age'].apply(lambda x : age_map[x])
```

```python
# col 생성
df2 = ofn_age_cnt.drop_duplicates(subset=['jur_stn','date'])[['jur_stn','date']]
df2= df2.reset_index()
del df2['index']
for i in ofn_age_cnt['ofn_age'].unique() :
    df2['ofn_age_'+i] = 0
df2.head(3)
```

```python
for i in range(len(ofn_age_cnt)) :
    age_name = ofn_age_cnt['ofn_age'].iloc[i]
    temp = df2[(df2['jur_stn']==ofn_age_cnt['jur_stn'].iloc[i])&(df2['date']==ofn_age_cnt['date'].iloc[i])]
    temp[age_name] = ofn_age_cnt['ofn_age_cnt'].iloc[i]
    df2['ofn_age_'+age_name].iloc[temp.index] = ofn_age_cnt['ofn_age_cnt'].iloc[i]
df2.head(3)
```

```python
## 가해자 df merge
ofn_df = pd.merge(df1,df2, on=['jur_stn', 'date'])
ofn_df.head(3)
```

```python
ofn_df['date'].value_counts()
```

### 피해자

```python
# 피해자 성별 cnt
vic_sx_cnt = car_acc.groupby(['jur_stn','date','vic_sx']).count().unstack(fill_value=0).stack()[['acc_case']]
vic_sx_cnt.reset_index(inplace=True)
vic_sx_cnt = vic_sx_cnt.rename(columns={'acc_case' : 'vic_sx_cnt'})
vic_sx_cnt.head()
```

```python
# col 생성
df1 = vic_sx_cnt.drop_duplicates(subset=['jur_stn','date'])[['jur_stn','date']]
df1= df1.reset_index()
del df1['index']
for i in vic_sx_cnt['vic_sx'].unique() :
    df1['vic_sx_'+i] = 0
df1.head(3)
```

```python
for i in range(len(vic_sx_cnt)) :
    sx_name = vic_sx_cnt['vic_sx'].iloc[i]
    temp = df1[(df1['jur_stn']==vic_sx_cnt['jur_stn'].iloc[i])&(df1['date']==vic_sx_cnt['date'].iloc[i])]
    df1['vic_sx_'+sx_name].iloc[temp.index] = vic_sx_cnt['vic_sx_cnt'].iloc[i]
df1.head(3)
```

```python
# 피해자 성비
df1['vic_sx_성비'] = round(df1['vic_sx_남'] / df1['vic_sx_여'] * 100, 1)
df1.head(3)
```

```python
## 피해자 나이 cnt
vic_age_cnt = car_acc.groupby(['jur_stn','date','vic_age']).count().unstack(fill_value=0).stack()[['acc_case']]
vic_age_cnt.reset_index(inplace=True)
vic_age_cnt = vic_age_cnt.rename(columns={'acc_case' : 'vic_age_cnt'})
vic_age_cnt.head()
```

```python
vic_age_cnt['vic_age'] = vic_age_cnt['vic_age'].apply(lambda x : age_map[x]) # age 재분류
```

```python
df2 = vic_age_cnt.drop_duplicates(subset=['jur_stn','date'])[['jur_stn','date']]
df2= df2.reset_index()
del df2['index']
for i in vic_age_cnt['vic_age'].unique() :
    df2['vic_age_'+i] = 0
df2.head(3)
```

```python
for i in range(len(vic_age_cnt)) :
    age_name = vic_age_cnt['vic_age'].iloc[i]
    temp = df2[(df2['jur_stn']==vic_age_cnt['jur_stn'].iloc[i])&(df2['date']==vic_age_cnt['date'].iloc[i])]
    temp[age_name] = vic_age_cnt['vic_age_cnt'].iloc[i]
    df2['vic_age_'+age_name].iloc[temp.index] = vic_age_cnt['vic_age_cnt'].iloc[i]
df2.head(3)
```

```python
vic_df = pd.merge(df1,df2, on=['jur_stn', 'date'])
vic_df.head(3)
```

```python
vic_df['date'].value_counts()
```

### 월 교통사고 건수

```python
# 상반기 1 ~ 6월 , 하반기 7 ~ 12월로 분류하여 적용
month = {"1월" : "1m", "2월" : "2m", "3월" : "3m", "4월" : "4m", "5월" : "5m", "6월" : "6m", 
         "7월" : "1m", "8월" : "2m" , "9월" : "3m" , "10월" : "4m" , "11월" : "5m", "12월" : "6m"
        }
car_acc['half_month'] = car_acc['month'].map(lambda x : month[x])
car_acc.head(3)
```

```python
month_cnt = car_acc.groupby(['jur_stn','date','half_month']).count().unstack(fill_value=0).stack()[['acc_case']]
month_cnt.reset_index(inplace=True)
month_cnt = month_cnt.rename(columns={'acc_case' : 'month_cnt'})
month_cnt.head()
```

```python
# col 생성
month_df = month_cnt.drop_duplicates(subset=['jur_stn','date'])[['jur_stn','date']]
month_df= month_df.reset_index()
del month_df['index']
for i in month_cnt['half_month'].unique() :
    month_df['month_'+i] = 0
month_df.head(3)
```

```python
for i in range(len(month_cnt)) :
    month_name = month_cnt['half_month'].iloc[i]
    temp = month_df[(month_df['jur_stn']==month_cnt['jur_stn'].iloc[i])&(month_df['date']==month_cnt['date'].iloc[i])]
    month_df['month_'+month_name].iloc[temp.index] = month_cnt['month_cnt'].iloc[i]
month_df.head(3)
```

### act_type

```python
# act type cnt
type_cnt = car_acc.groupby(['jur_stn','date','acc_type']).count().unstack(fill_value=0).stack()[['acc_case']]
type_cnt.reset_index(inplace=True)
type_cnt = type_cnt.rename(columns={'acc_case' : 'acc_type_cnt'})
type_cnt.head()
```

```python
# col 생성
df1 = type_cnt.drop_duplicates(subset=['jur_stn','date'])[['jur_stn','date']]
df1= df1.reset_index()
del df1['index']
for i in type_cnt['acc_type'].unique() :
    df1['acc_type_'+i] = 0
df1.head(3)
```

```python
for i in range(len(type_cnt)) :
    type_name = type_cnt['acc_type'].iloc[i]
    temp = df1[(df1['jur_stn']==type_cnt['jur_stn'].iloc[i])&(df1['date']==type_cnt['date'].iloc[i])]
    df1['acc_type_'+type_name].iloc[temp.index] = type_cnt['acc_type_cnt'].iloc[i]
df1.head(3)
```

### acc_case

```python
# acc_case cnt
case_cnt = car_acc.groupby(['jur_stn','date','acc_case']).count().unstack(fill_value=0).stack()[['ofn_sx']]
case_cnt.reset_index(inplace=True)
case_cnt = case_cnt.rename(columns={'ofn_sx' : 'acc_case_cnt'})
case_cnt.head()
```

```python
df2 = case_cnt.drop_duplicates(subset=['jur_stn','date'])[['jur_stn','date']]
df2 = df2.reset_index()
del df2['index']

for i in case_cnt['acc_case'].unique() :
    df2['acc_case_'+i] = 0
df2.head(3)
```

```python
for i in range(len(case_cnt)) :
    case_name = case_cnt['acc_case'].iloc[i]
    temp = df2[(df2['jur_stn']==case_cnt['jur_stn'].iloc[i])&(df2['date']==case_cnt['date'].iloc[i])]
    df2['acc_case_'+case_name].iloc[temp.index] = case_cnt['acc_case_cnt'].iloc[i]
df2.head(3)
```

```python
### act_type, act_case Merge
acc_df = pd.merge(df1,df2, on=['jur_stn', 'date'])
acc_df.head(3)
```

```python
acc_df['date'].value_counts()
```
