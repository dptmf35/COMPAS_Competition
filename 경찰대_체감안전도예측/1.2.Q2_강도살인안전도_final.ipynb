## ※ 목표 : Q2 X값 생성

- 전체 112 신고건과 강도, 살인 관련 신고건을 count하여 x값으로 사용

```python
report_112.head(3) # Q2에 사용할 data 
```

### 1. 강도

```python
# 강도 total count
robb_total_cnt = report_112.groupby(['jur_stn','date','inc_info']).count().unstack(fill_value=0).stack()[['report_sx']]
robb_total_cnt.reset_index(inplace=True)
robb_total_cnt = robb_total_cnt.rename(columns={'report_sx' : 'robb_total_cnt'})
robb_total_cnt = robb_total_cnt[robb_total_cnt['inc_info']=='강도']
robb_total_cnt.head()
```

```python
robb_total_cnt['robb_total_cnt'] = robb_total_cnt.apply(lambda x : x['robb_total_cnt'] * 5 if x['date'] =='2018(상반기)' else x['robb_total_cnt'] , axis= 1   )
# 0 일경우 1로 대체
robb_total_cnt['robb_total_cnt'] = robb_total_cnt.apply(lambda x : 1 if x['robb_total_cnt'] ==0 else x['robb_total_cnt'] , axis= 1   )

del robb_total_cnt['inc_info']
robb_total_cnt.head(3)
```

### 2. 살인

```python
# 살인 신고자 total count
mur_total_cnt = report_112.groupby(['jur_stn','date','inc_info']).count().unstack(fill_value=0).stack()[['year']]
mur_total_cnt.reset_index(inplace=True)
mur_total_cnt = mur_total_cnt.rename(columns={'year' : 'mur_total_cnt'})
mur_total_cnt = mur_total_cnt[mur_total_cnt['inc_info']=='살인']
mur_total_cnt.head()
```

```python
mur_total_cnt['mur_total_cnt'] = mur_total_cnt.apply(lambda x : x['mur_total_cnt'] * 5 if x['date'] =='2018(상반기)' else x['mur_total_cnt'] , axis= 1   )
# 0 일경우 1로 대체
mur_total_cnt['mur_total_cnt'] = mur_total_cnt.apply(lambda x : 1 if x['mur_total_cnt'] ==0 else x['mur_total_cnt'] , axis= 1   )
```

### 1, 2 merge 

```python
# 강도 + 살인 merge
temp_df = pd.merge(robb_total_cnt, mur_total_cnt)
temp_df.head()
```

## 3. X값 최종 Merge

```python
# feature로 사용될 데이터 종합
q2_df = pd.merge( temp_df, q2_info, how='outer')
q2_df.head()
```

```python
del q2_df['inc_info']
```

```python
q2_df['date'].value_counts()
```

```python
q2_df[q2_df['jur_stn'].str.contains('세종')]
```

```python
## 세종 null 값 index 채우기 ##
null_index= q2_df[q2_df['cctv_cnt'].isnull()].index 
sejong_mean = q2_df[(q2_df['jur_stn'].str.contains('세종')) & (q2_df['robb_total_cnt'].notnull()) ].mean()
```

```python
for i in null_index :
    for c in q2_df.columns[2:-1] :
#         print(i, c)
#         print(q1_total.loc[i, c])
        q2_df.loc[i, c] = sejong_mean[c]
```

```python
q2_df['robb_total_cnt'].fillna(1, inplace=True)
q2_df['mur_total_cnt'].fillna(1, inplace=True)
```

```python
q2_df[q2_df['jur_stn'].str.contains('세종')]
```
